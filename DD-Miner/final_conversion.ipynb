```python
import pandas as pd
```

```python
doid2mesh = pd.read_csv("doid_to_mesh.csv")
```

```python
doid2mesh.head()
```

```python
dd_miner = pd.read_csv("DD-Miner_miner-disease-disease.tsv", sep='\t')
```

```python
dd_miner.head()
```

```python
first_doid2mesh = pd.merge(dd_miner, doid2mesh, left_on="# Disease(DOID)", right_on="DOID")
```

```python
first_doid2mesh.rename(columns={"MESH":"MESH_1"}, inplace=True)
del first_doid2mesh["DOID"]
```

```python
second_doid2mesh = pd.merge(first_doid2mesh, doid2mesh, left_on="Disease(DOID)", right_on="DOID")
```

```python
second_doid2mesh.rename(columns={"MESH":"MESH_2"}, inplace=True)
del second_doid2mesh["# Disease(DOID)"]
del second_doid2mesh["Disease(DOID)"]
del second_doid2mesh["DOID"]
```

```python
second_doid2mesh
```

```python
second_doid2mesh = second_doid2mesh.drop_duplicates()
```

```python
second_doid2mesh.to_csv("disease_disease.csv", index=False)
```
